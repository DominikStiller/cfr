# Get the nearest climate for proxies

In this tutorial, we introduce how to get the nearest climate for proxies with `cfr`.

```python
%load_ext autoreload
%autoreload 2

import cfr
import pandas as pd
import numpy as np
import os

import warnings
warnings.filterwarnings('ignore')
```

## Load the PAGES2k network stored in a pandas `DataFrame`

```python
df = pd.read_pickle('./data/pages2k_dataset.pkl')
df.columns
```

## Create a `ProxyDatabase` and load from the pandas `DataFrame`

```python
pdb = cfr.ProxyDatabase().from_df(df)
```

## Load climate data to `ClimateDataset`

```python
dirpath = './data'

ds_tas = cfr.ClimateDataset().load_nc(os.path.join(dirpath, 'tas_sfc_Amon_iCESM_past1000historical_085001-200512.nc'))
ds_pr = cfr.ClimateDataset().load_nc(os.path.join(dirpath, 'pr_sfc_Amon_iCESM_past1000historical_085001-200512.nc'))
```

## Get the nearest climate from a `ClimateField` or a `ClimateDataset` for a `ProxyRecord`

```python
fd = ds_tas.fields['tas']
```

```python
pobj = pdb.records['Ant_005']
fig, ax = pobj.plot()
```

```python
pobj.get_clim(fd)
```

```python
pobj.get_clim(ds_pr)
```

```python
pobj.clim['tas'].da
```

```python
pobj.clim['pr'].da
```

## Get the nearest climate from a `ClimateField` for a `ProxyDatabase`

```python
pdb_TRW = pdb.filter(by='ptype', keys=['tree.TRW'])
```

```python
pdb_TRW.del_clim()
pdb_TRW.get_clim(ds_tas+ds_pr, tag='model')
```

```python
pdb_TRW.records['NAm_153'].clim['model-tas'].da
```

```python
pdb_TRW.records['NAm_153'].clim['model-pr'].da
```

```python
pdb_TRW.records['NAm_153'].clim.keys()
```

```python
pd.to_pickle(pdb_TRW, './data/pdb_TRW_clim.pkl')
```
