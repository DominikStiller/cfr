# Analysis

In this tutorial, we introduce the basic analysis of climate datasets with `cfr`.

```python
%load_ext autoreload
%autoreload 2

import cfr
import pandas as pd
import numpy as np
import os


import warnings
warnings.filterwarnings('ignore')
```

## Load test data to `ClimateDataset`

```python
dirpath = '/Volumes/ExtraDrive/Data/GCM_sims/icesm_last_millennium_historical'

ds_tas = cfr.ClimateDataset().load_nc(os.path.join(dirpath, 'tas_sfc_Amon_iCESM_past1000historical_085001-200512.nc'))
ds_pr = cfr.ClimateDataset().load_nc(os.path.join(dirpath, 'pr_sfc_Amon_iCESM_past1000historical_085001-200512.nc'))
```

## Get the anomaly from a `ClimateField`

```python
fd_tas = ds_tas.fields['tas']
fd_tas_anom = fd_tas.get_anom(ref_period=[1951, 1980])
fig, ax = fd_tas_anom.plot(it=-1)
```

## Get the anomly from a `ClimateDataset`

```python
ds_tas_anom = ds_tas.get_anom(ref_period=[1951, 1980])
```

```python
fig, ax = ds_tas_anom.fields['tas'].plot(it=-1)
```

## Annualize a `ClimateField`

```python
fd_tas = ds_tas.fields['tas']
fd_tas.time
```

```python
da_ann = cfr.utils.annualize(da=fd_tas.da)
```

```python
print(da_ann.time)
```

## Annualize a `ClimateDataset`

```python
ds_all = ds_tas + ds_pr
print(ds_all.nv)
```

```python
print(ds_all.fields['tas'].time)
```

```python
ds_ann = ds_all.annualize()
```

```python
ds_ann.ds
```

```python
print(ds_ann.fields['tas'].time)
fig, ax = ds_ann.fields['tas'].plot(it=-1)
```

```python
ds_djf = ds_all.annualize(months=[12, 1, 2])
```

```python
print(ds_djf.fields['tas'].time)
fig, ax = ds_djf.fields['tas'].plot(it=-1)
```

```python
fig, ax = ds_djf.fields['pr'].plot(it=-1)
```

## Regrid a `ClimateField`

```python
fd_tas = ds_djf.fields['tas']
fd_tas.da.coords
```

```python
fig, ax = fd_tas.plot(it=-1)
```

```python
fd_regrid = fd_tas.regrid(np.arange(-90, 90.1, 5), np.arange(0, 360, 5))
fd_regrid.da.coords
```

```python
fig, ax = fd_regrid.plot(it=-1)
```
